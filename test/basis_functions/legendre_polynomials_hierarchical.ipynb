# Test Hierarchical Legendre Polynomials

```julia
using GeometricIntegrators
```

```julia
using Gadfly
```

Number of basis functions to plot:

```julia
s = 5
```

Number of points to plot:

```julia
np = 31
```

Define Lagrange polynomial:

```julia
leg = LegendreBasisHierarchical(Float64, s)
```

Compute plot range:

```julia
xp = collect(linspace(0, 1, np));
```

Plot all basis polynomials:

```julia
layers = []

for i in 1:nbasis(leg)
    yp = zeros(xp)
    for j in eachindex(xp,yp)
        yp[j] = evaluate(leg, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```

```julia
layers = []

for i in 1:nbasis(leg)
    yp = zeros(xp)
    for j in eachindex(xp,yp)
        yp[j] = derivative(leg, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```

```julia
layers = []

for i in 1:nbasis(leg)
    yp = zeros(xp)
    for j in eachindex(xp,yp)
        yp[j] = integral(leg, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```
